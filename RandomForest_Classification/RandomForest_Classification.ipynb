```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the CSV file
data = pd.read_csv("Social_Network_Ads.csv")
data.head(5)
```

```python
#Print the shape of data
print("No of rows :",data.shape[0])
print("No of columns",data.shape[1])
```

```python
#Get the info of data:
data.info()
```

```python
#Delete the unwanted feature:
data = data.drop(['User ID'], axis=1)
data.head(5)
```

```python
#Since we have one object value ('Gender') - convert to numerical value
df = pd.get_dummies(data,drop_first = True, dtype=int)
df.head(5)
```

```python
# split the data into X and Y
independent = df.drop(columns=['Purchased'])
dependent = df['Purchased']
```

```python
independent.head(3)
```

```python
#Get the count of each value in parchesed feature
dependent.value_counts()
```

```python
dependent.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=150,criterion='gini',max_features='log2',verbose=3,n_jobs=-1)
rf_classifier.fit(X_train,y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
cm_result = confusion_matrix(y_test,y_pred)
cm_result
```

```python
cr_report = classification_report(y_test,y_pred)
```
